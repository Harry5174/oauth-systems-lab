# OAuth & CIMD Invariants

This notebook documents the non-negotiable security invariants
behind OAuth 2.0 and Client-ID Metadata Documents (CIMD).

Each invariant is:
- explained conceptually
- illustrated with a concrete example
- broken intentionally|
- restated as a one-line rule


## Invariant 1 — client_id must be HTTPS

The client_id represents identity.
If identity is fetched over an insecure channel,
the entire system collapses.

client_id = URL

URL without HTTPS = untrusted identity


```python
def validate_client_id(client_id: str):
    assert client_id.startswith("https://")
```

```python
g_str = "https://example.com"
b_str = "http://example.com"

validate_client_id(g_str)
validate_client_id(b_str)
```
